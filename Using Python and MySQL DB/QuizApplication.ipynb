```python
import random
def getQuestions():
    questions=random.choice(questionsStore)
    print(questions[0])
    print("1. ",questions[1])
    print("2. ", questions[2])
    print("3. ",questions[3])
    print("4. ",questions[4])
    correctAnswer=questions[5]
    userAnswer=int(input())
    if userAnswer==correctAnswer:
        print("Correct Answer!!!")
        print()
        return True
    print("Wrong Answer!!!")
    print("Correct answer is Option",correctAnswer,".",questions[correctAnswer])
    print()
    return False
```

```python
option="Y"
score=0
count=0
name=input("Enter your name: ")
print("Welcome",name)
print("The rules are very simple. You get +3 for every correct answer and -1 for wrong answers")
print("Let's begin!")
print()
while True:
    result=getQuestions()
    if result==True:
        score += 3
    else:
        score += -1
    
    count += 1
    if count%5==0:
        option=input("Do you want to continue playing? Y/N ")
    if option.lower()=="n" or count==dataCount:
        break
print()
print("Thank You for taking the Quiz,",name)
print()
print("Result:")
if score==1:
    print("You scored",score,"point out of",count*3,"points")
else:
    print("You scored",score,"points out of",count*3,"points")
```
